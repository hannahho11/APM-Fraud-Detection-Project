```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

training = pd.read_csv('train_transaction.csv')
```

```python
# consider only columns with less than 1% NaN
low_null_columns = (training.isnull().sum()/len(training)*100) < 1
low_nulls = []
index=0
for column in low_null_columns:
    if column == True:
        low_null_columns.index[index]
        low_nulls.append(low_null_columns.index[index])
    index += 1
training_imputed = training.loc[:, low_nulls]
training_imputed = training_imputed.dropna()
training_imputed.shape
```
