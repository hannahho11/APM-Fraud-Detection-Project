```python
#print(df_train[df_train['dist1'] > 1000]['isFraud'].value_counts())
def fraud_percentage(fraud_val_counts):
    if len(fraud_val_counts) > 1:
        return 1.0*fraud_val_counts[1]/(sum(fraud_val_counts))
    elif fraud_val_counts[0] == 0:
        return 0
    else:
        return 1

fraud_percentage(df_train[df_train['dist1'] > 2000]['isFraud'].value_counts())
```

```python
print('Fraud Rate per product\n')
for code in df_train['ProductCD'].unique():
    print(code,fraud_percentage(df_train[df_train['ProductCD'] == code]['isFraud'].value_counts()))
print('\nProducts Sold\n')
print(df_train['ProductCD'].value_counts())
print(df_train[''])
```
